<b><h1>Anomaly detection  using OneClassSVM</h1></b>

OneClassSVM

  1.   is an unsupervised algorithm for anomaly detection.
  2.   learns the boundary for the normal data points and identifies the data outside the border to be anomalies.
  3.  labels the normal data 1 and anomalies  -1.

To undrestand the concept of this  algorithm and how it works  , we will try to detect  the abnormal point data from the iris dataset.





```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.svm import OneClassSVM
```

```python
#load the iris dataset
iris_data = datasets.load_iris()
```

```python
# we only take the first two features
df = iris_data.data[: , :2] 
```

```python
#plot the data 
plt.figure(figsize = (8,6))
plt.scatter(df[:,0] ,df[:,1])
plt.xlabel(iris_data.feature_names[0])
plt.ylabel(iris_data.feature_names[1])
plt.title('Iris dataset')
plt.show()
```

```python
# function to detect the abnormal iris
def model(df , per = 0.05):

  # nu :the estimated proportion of anomalies expected  in this data. 
  model = OneClassSVM(kernel='rbf', gamma=0.001 , nu=per)
  # train the model
  model.fit(df)
  # prediction
  pred = model.predict(df)

  return pred
```

```python
pred1 = model(df)
anomaly_values1 = df[np.where(pred1 == -1)]
```

```python
#plot the data 
plt.figure(figsize = (8,6))
plt.scatter(df[:,0] ,df[:,1])
plt.scatter(anomaly_values1 [:,0] ,anomaly_values1 [:,1] , c='r')
plt.xlabel(iris_data.feature_names[0])
plt.ylabel(iris_data.feature_names[1])
plt.title('Anomaly Iris detection ')
plt.show()
```

```python
per = 0.1
pred2 = model(df , per)
anomaly_values2 = df[np.where(pred2 == -1)]
```

```python
# with 5% of anomalous data
plt.figure(figsize = (20,8))
plt.subplot(1,2,1)
plt.scatter(df[:,0] ,df[:,1])
plt.scatter(anomaly_values1 [:,0] ,anomaly_values1 [:,1] , c='r')
plt.xlabel(iris_data.feature_names[0])
plt.ylabel(iris_data.feature_names[1])
plt.title('with 5% of anomalous data  ')

#with 10% of anomalous data 
plt.subplot(1,2,2)
plt.scatter(df[:,0] ,df[:,1])
plt.scatter(anomaly_values2 [:,0] ,anomaly_values2 [:,1] , c='r')
plt.xlabel(iris_data.feature_names[0])
plt.ylabel(iris_data.feature_names[1])
plt.title('with 10% of anomalous data ')
plt.show()
```
